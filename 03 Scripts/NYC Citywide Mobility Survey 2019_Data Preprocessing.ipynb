## Import data

```python
# Define path
path = r'/Users/peimei/11-2023 New York City Citywide Mobility Survey 2019'

# Join the path using 'os.path.join'
file_path = os.path.join(path, '02 Data', 'Original Data', 'Citywide_Mobility_Survey_-_Trip_Survey_2019_20231108.csv')

# Read the CSV file into a DataFrame
trips = pd.read_csv(file_path, sep=";", index_col = False,low_memory=False)

# Display the first few rows of the DataFrame
trips.head()
```

## Check for mixed-type data

```python
for col in trips.columns.tolist():
  weird = (trips[[col]].applymap(type) != trips[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (trips[weird]) > 0:
    print (col)
```

```python
# Modify data types of the following variables respectively

trips['trip_weight'] = trips['trip_weight'].str.replace(',', '').astype('float64')
trips['o_cms_zone'] = trips['o_cms_zone'].astype('str')
trips['d_cms_zone'] = trips['d_cms_zone'].astype('str')
trips['mode_1'] = trips['mode_1'].str.replace(',', '').astype('int64')
trips['trip_distance'] = trips['trip_distance'].str.replace(',', '').astype('float64')
trips['trip_duration'] = trips['trip_duration'].str.replace(',', '').astype('float64')
trips['d_distance_home'] = trips['d_distance_home'].str.replace(',', '').astype('float64')
trips['d_distance_work'] = trips['d_distance_work'].str.replace(',', '').astype('float64')
trips['d_distance_school'] = trips['d_distance_school'].str.replace(',', '').astype('float64')
```

```python
# Check output

for col in trips.columns.tolist():
  weird = (trips[[col]].applymap(type) != trips[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (trips[weird]) > 0:
    print (col)
else:
        print('All good.')
```

## Remove outliers

```python
# Check how many trips last zero minutes
trips_duration_zero = trips_trimmed[trips_trimmed['trip_duration_min'] == 0]
trips_duration_zero.shape[0]
```

```python
# Check how many trips last more than 24 hours
trips_more_than_24hr = trips_trimmed[trips_trimmed['trip_duration_min'] > 24*60]
trips_more_than_24hr.shape[0]
```

```python
# Check how many trips are zero miles long
trips_distance_zero = trips_trimmed[trips_trimmed['trip_distance_mile'] == 0]
trips_distance_zero.shape[0]
```

```python
# Assume the longest travel distance by any transportation is 1000 miles a day.
# Check how many trips are longer than 1000 miles
trips_more_than_1000miles = trips_trimmed[trips_trimmed['trip_distance_mile'] > 1000]
```

```python
trips_more_than_1000miles.shape[0]
```

### As the sample size won't be significantly reduced by applying these four conditions, I decided to <br/>1) remove outliers by limiting the trip duration to 1440 minutes & the trip distance to 1000 miles and <br/>2) remove zeros in both columns.

```python
trips_trimmed_2 = trips_trimmed[(trips_trimmed['trip_duration_min'] <= 24*60) & 
    (trips_trimmed['trip_duration_min'] != 0) & (trips_trimmed['trip_distance_mile'] <= 1000)]
```
